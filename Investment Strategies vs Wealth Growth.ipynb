```python
# How different Investments affect Patrimonial Evolution

# Initial parameters
starting_capital = 10000 # euros
monthly_contributions = 100 # euros
years = 15
months = years * 12

# Annual interest rates (realistic examples)
interest_rates = {
    'no_investment': 0.00,
    'bank_savings': 0.01,       # e.g., 1% annual
    'conservative_fund': 0.035  # e.g., 3.5% annual  
}

# Convert annual interest rates to monthly
monthly_rates = {k: v / 12 for k, v in interest_rates.items()}
```

```python
import pandas as pd

# Create initial month column
df = pd.DataFrame({'Month': range(1, months + 1)})

# Loop through each strategy
for strategy, rate in monthly_rates.items():
    balance = []
    total = starting_capital

    for month in range(months):
        if rate > 0:
            total *= (1 + rate)
        total += monthly_contribution
        balance.append(total)

    df[strategy] = [round(val, 2) for val in balance]
```

```python
import matplotlib.pyplot as plt

# Set up the plot
plt.figure(figsize=(12, 6))

# Plot each investment strategy
for column in df.columns[1:]: # Skip 'Month'
    plt.plot(df['Month'], df[column], label=column.replace('_', ' ').title())

# Customize the chart
plt.title('Investment Growth Over Time')
plt.xlabel('Month')
plt.ylabel('Total Value ()')
plt.legend()
plt.grid(True)

# Highlight final values with dotted vertical lines + labels
final_month = df['Month'].iloc[-1]
for column in df.columns[1:]:
    final_value = df[column].iloc[-1]
    plt.axvline(final_month, color='red', linestyle='--', linewidth=0.8)
    plt.text(final_month + 1, final_value, f"{final_value:,.2f}", va='center', fontsize=11)

# Increase graph size
plt.xlim(0, final_month + 18)

# Show the plot 
plt.tight_layout()
plt.show()
```

# Final Observations

This simulation compares how an individual's wealth evolves over 10 years under three different scenarios, assuming monthly contributions of 100:

    - No Investment: Pure savings without any interest - capital grows only through contributions.
    - Bank Savings: A conservative option with low annual return (1%).
    
    - Conservative Fund: A modest investment option with a higher annual return (3.5%).
    
> *Note: This simulation does not consider inflation. While real returns would be slightly lower, the relative comparison remains valid in a low-inflation environment like the Eurozone.*

It is important to note that this simulation is based on a European scenario, where interest rates tend to be low. The investment types chosen were among the most conservative available. 
This shows that even small differences in interest rates can significantly impact overall wealth growth over the medium to long term.
